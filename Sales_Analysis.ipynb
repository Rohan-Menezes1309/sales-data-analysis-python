```python
columns_needed= ["Order.Date","Ship.Date","Region","Category","Sub.Category","Sales","Profit","Quantity","Discount","Segment","Shipping.Cost","Year"]
df= df[columns_needed]
df.head()
```

```python
df["Order.Date"]= pd.to_datetime(df["Order.Date"])
df["Ship.Date"]= pd.to_datetime(df["Ship.Date"])
df.dtypes
```

```python
df["Month"]= df["Order.Date"].dt.month
df.head()
```

```python
df.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

region_sales= df.groupby("Region")["Sales"].sum()
sns.barplot(x=region_sales.index, y=region_sales.values)
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.savefig("Total_Sales_by_region")
plt.show()
```

Total sales vary significantly across regions, indicating that some regions contribute substantially more revenue than others.

```python
category_profit= df.groupby("Category")["Profit"].sum()
sns.barplot(x=category_profit.index, y=category_profit.values)
plt.title("Total Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profits")
plt.xticks(rotation=45)
plt.show()
plt.savefig("Total_profit_by_category")
```

The Central region generated the highest total sales, indicating it is the strongest revenue-contributing region.

Canada recorded the lowest total sales, suggesting relatively low market penetration or demand in that region.

Technology products generated the highest total profit, making it the most financially valuable product category.

Furniture generated the lowest profit among categories, suggesting lower profit margins or higher associated costs.

```python
monthly_sales= df.groupby("Month")["Sales"].sum()
plt.figure(figsize=(10,5))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
plt.savefig("Monthly_Sales")
```

Sales peak in December, indicating strong year-end demand, likely driven by holiday or seasonal purchasing patterns.

Sales are lowest in February, suggesting reduced consumer activity following the peak sales period.

Overall sales show an increasing trend toward the end of the year, indicating seasonal demand concentration in later months.

```python
subcategory_sales= df.groupby("Sub.Category")["Sales"].sum().sort_values(ascending= False)
plt.figure(figsize=(12,6))
sns.barplot(x= subcategory_sales.index, y= subcategory_sales.values)
plt.title("Sales by Sub Category")
plt.xlabel("Sub Category")
plt.ylabel("Total Sales")
plt.show()
plt.savefig("Sales_by_Subcategory")
```

Phones generated the highest total sales among all sub-categories, indicating strong market demand for communication-related products.


Copiers ranked as the second highest revenue-generating sub-category, suggesting consistent demand for office-related equipment.

Labels recorded the lowest sales, indicating limited demand or niche usage compared to other product groups.

```python
plt.figure(figsize=(8,5))
sns.regplot(x="Discount", y="Profit", data=df, scatter_kws={"alpha":0.3})
plt.title("Discount v/s Profit Relationship")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.show()
plt.savefig("discount_vs_profit")
```

The analysis shows that higher discount levels are associated with lower profitability. A significant number of transactions with high discounts result in negative profit, indicating that excessive discounting can lead to losses.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["Shipping.Cost"], y=df["Profit"])
plt.title("Shipping Cost vs Profit")
plt.xlabel("Shipping Cost")
plt.ylabel("Profit")
plt.show()
plt.savefig("shipping_cost_vs_profit")
```

The relationship between shipping cost and profit appears weak, indicating that higher shipping costs do not necessarily lead to increased profitability.

Final Summary: This project analysed sales and profitability data to identify key business trends and performance drivers.

Key findings include:
1. The Central region contributed the highest sales, while Canada had the lowest.
2. Technology was the most profitable category, whereas Furniture generated the lowest profit.
3. Sales showed strong seasonal trends, peaking in December and dipping in February.

6. Shipping cost showed little to no strong relationship with profit.
